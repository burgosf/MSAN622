```python
SITE = "http://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
START = datetime(2015, 4, 30, 0, 0, 0, 0, pytz.utc)
END = datetime(2015, 5, 9, 0, 0, 0, 0, pytz.utc)
```

```python
def scrape_list(site):
    hdr = {'User-Agent': 'Mozilla/5.0'}
    req = urllib2.Request(site, headers=hdr)
    page = urllib2.urlopen(req)
    soup = BeautifulSoup(page)

    table = soup.find('table', {'class': 'wikitable sortable'})
    sector_tickers = dict()
    for row in table.findAll('tr'):
        col = row.findAll('td')
        if len(col) > 0:
            sector = str(col[3].string.strip()).lower().replace(' ', '_')
            ticker = str(col[0].string.strip())
            if sector not in sector_tickers:
                sector_tickers[sector] = list()
            sector_tickers[sector].append(ticker)
    return sector_tickers
```

```python
sector_tickers = scrape_list(SITE)
```

```python
linechart = pd.DataFrame(columns=['Date', 'information_technology', 'industrials', 'energy', 'consumer_discretionary', 'utilities', 'materials', 'telecommunications_services', 'consumer_staples', 'financials', 'health_care'])
linechart.Date = ['2015-05-01', '2015-05-04', '2015-05-05', '2015-05-06', '2015-05-07', '2015-05-08']
```

```python
#
# Simulating Geometric Brownian Motion with Python
#
import math
import numpy as np
from random import gauss

# Parameters
S0 = 100; # current bitcoin price
mu = np.array([0.13, 0.083, 0.183, 0.168, 0.11, 0.118, 0.06, 0.09, 0.078, 0.085]); 
sigma = np.array([0.154, 0.175, 0.112, 0.15, 0.163, 0.103, 0.193, 0.186, 0.12, 0.133]); 
T = 1.0; # 1 Time cycle
M = 5; # 5 steps = 1 week
dt = T / 260 # dt
```

```python
import numpy as np
def genS_np(I, i):
    S = S0 * np.exp(np.cumsum((mu[i] - 0.5 * sigma[i] ** 2) * dt + sigma[i] * np.sqrt(dt) * np.random.standard_normal((M + 1, I)), axis=0))
    S[0] = S0
    return S
```

```python
linechart.information_technology = np.around(genS_np(1, 0), decimals=2)
linechart.financials = np.around(genS_np(1, 1), decimals=2)
linechart.health_care = np.around(genS_np(1, 2), decimals=2)
linechart.consumer_discretionary = np.around(genS_np(1, 3), decimals=2)
linechart.industrials = np.around(genS_np(1, 4), decimals=2)
linechart.consumer_staples = np.around(genS_np(1, 5), decimals=2)
linechart.energy = np.around(genS_np(1, 6), decimals=2)
linechart.materials = np.around(genS_np(1, 7), decimals=2)
linechart.utilities = np.around(genS_np(1, 8), decimals=2)
linechart.telecommunications_services = np.around(genS_np(1, 9), decimals=2)
```

```python
maxPerf = pd.DataFrame(columns=['sector', 'stock', 'best_return'])
sectoriel = list()
share = list()
best_return = list()

for secteur in secteurs:
#     print secteur
    for action in secteurs[secteur]:
#         print action
        tableau = np.array(secteurs[secteur][action])
#         print tableau
        maxi = -1000.0
        for i in range(len(tableau)-1):
            maxi = max(maxi, float(tableau[i+1]) / float(tableau[i]) - 1.0)
        sectoriel.append(secteur)
        share.append(action)
        best_return.append(maxi)

maxPerf['sector'] = sectoriel
maxPerf['stock'] = share
maxPerf['best_return'] = best_return
            
```
