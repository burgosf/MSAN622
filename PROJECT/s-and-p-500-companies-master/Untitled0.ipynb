```python
flare = dict()
flare["name"] = "flare"
flare["children"] = list()
for sect in list(data.Sector.unique()):
    sector = dict()
    sector['name'] = sect
    sector['children'] = list()
    for i in range(0, 496):
        if data.loc[i, 'Sector'] == sect:
            sector['children'].append({"name":data.loc[i, 'Name'], "size":round(data.loc[i, 'Market Cap'], 2)})
    flare["children"].append(sector)
```

```python
flare
```

```python
with open(path+'flare.json', 'w') as outfile:
    json.dump(flare, outfile)
```
