```python
#
# Simulating Geometric Brownian Motion with Python
#
import math
import numpy as np
from random import gauss

# Parameters
S0 = 100; # current bitcoin price
mu = np.array([0.13, 0.083, 0.183, 0.168, 0.11, 0.118, 0.06, 0.09, 0.078, 0.085]); 
sigma = np.array([0.154, 0.175, 0.112, 0.15, 0.163, 0.103, 0.193, 0.186, 0.12, 0.133]); 
T = 1.0; # 1 Time cycle
M = 5; # 5 steps = 1 week
dt = T / 260 # dt
```

```python
# Simulating I paths with M time steps
def genS_py(I):
    ''' I : number of index level paths '''
    S = []
    for i in range(I):
        path = []
        for t in range(M + 1):
            if t == 0:
                path.append(S0)
            else:
                z = gauss(0.0, 1.0)
                St = path[t - 1] * math.exp((mu - 0.5 * sigma ** 2) * dt
                                      + sigma * math.sqrt(dt) * z)
                path.append(St)
        S.append(path)
    return S
```

```python
I = 100000
%time S = genS_py(I)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
for i in range(5):  # first five paths
    plt.plot(S[i])
plt.xlabel('Time Steps')
plt.ylabel('Index Level')
plt.grid(True)
```

```python
K = 1000.0
C0 = math.exp(-mu * T) * sum([max(path[-1] - K, 0) for path in S]) / I
round(C0, 3)
```

```python
import numpy as np
def genS_np(I, i):
    S = S0 * np.exp(np.cumsum((mu[i] - 0.5 * sigma[i] ** 2) * dt + sigma[i] * np.sqrt(dt) * np.random.standard_normal((M + 1, I)), axis=0))
    S[0] = S0
    return S
```

```python
%time S = genS_np(1, 1)
```
