```python
import pandas as pd
import numpy as np
import sys
```

```python
filepath = "/Users/flo/Desktop/MSAN/MSAN622/MSAN622/Homework2/"
```

```python
data=pd.read_csv(filepath+'statex77.csv')
print "data loaded"
print data.shape
```

```python
data["state.region"].unique()
```

```python
grouped = data.groupby(['state.division'])
```

```python
result = pd.DataFrame(columns=['state.division', 'Population', 'Income', 'Illiteracy', 'Life.Exp', 'Murder', 'HS.Grad', 'Frost', 'Area'])
for key, group in grouped:
    state_division = key
    Population = round(group.Population.mean(), 0)
    Income = round(group.Income.mean(), 0)
    Illiteracy = round(group.Illiteracy.mean(), 2)
    Life_Exp = round(group["Life.Exp"].mean(), 1)
    Murder = round(group.Murder.mean(), 2)
    HS_Grad = round(group["HS.Grad"].mean(), 2)
    Frost = round(group.Frost.mean().mean(), 1)
    Area = round(group.Area.mean().mean(), 0)
    result = result.append(pd.Series([state_division, Population, Income, Illiteracy, Life_Exp, Murder, HS_Grad, Frost, Area], index = ['state.division', 'Population', 'Income', 'Illiteracy', 'Life.Exp', 'Murder', 'HS.Grad', 'Frost', 'Area']), ignore_index=True)
```

```python
result
```

```python
result.to_csv(filepath+'small_multiple.csv',index=False)
```

```python
result = pd.DataFrame(columns=['features', 'Region', 'Number'])
features = ['Population', 'Income', 'Illiteracy', 'Life.Exp', 'Murder', 'HS.Grad', 'Frost', 'Area']
for f in features:
    grouped = data.groupby(['state.region'])
    for key, group in grouped:
#         print key
        state_region = key
#         print group
        number= round(group[f].mean(), 2)
        result = result.append(pd.Series([f, state_region, number], index = ['features', 'Region', 'Number']), ignore_index=True)
```

```python
result.to_csv(filepath+'small_multiple.csv',index=False)
```
